```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
x = np.arange(0,2*pi,0.1)
plt.plot(x,np.sin(x))
plt.title("Sine Function")
```

```python
plt.plot(x,np.cos(x))
plt.title("Cosine Function")
```

```python
plt.plot(x,np.tan(x))
plt.title("Tangent Function")
```

```python
plt.plot(x,np.tanh(x))
plt.title("Tanh Function")
```
