## MODUL PLOTLY

Import Modul Plotly

```python
import plotly.express as px
```

#Scater Plot

```python
import pandas as pd
# siapkan data set
df = pd.read_csv("mtcars.csv")
df["model"] = df["model"].astype("string")
df
```

```python
# Membuat Plot Scatler
figure = px.scatter(df, x = "mpg", y = "wt", color = "model", size = "cyl")
figure.update_layout(title='mtcars', title_font_color="blue",title_font_size=30,
                     xaxis=dict(title='mpg',
                                titlefont_size=16,
                                tickfont_size=14),
                     yaxis=dict(title='wt',
                                titlefont_size=16,
                                tickfont_size=14))
figure.show()
```

# Barplot


```python
import plotly.graph_objects as go
barplot = go.Figure()
barplot.add_trace(go.Scatter(x = "wt", name = "Konfirmasi", mode = "lines+markers"))

barplot.update_layout( 
    #Menambahkan dan mengatur Judul Grafis
    title='Diagram Garis COVID-19 Indonesia Hingga 17 Desember 2020', 
    title_font_color="blue",
    title_font_size=30, 
    #Menambahkan dan mengatur label sumbu x            
    xaxis=dict(
        title='Tanggal',
        titlefont_size=16,
        tickfont_size=14),
    #Menambahkan dan mengatur label sumbu y
    yaxis=dict(
        title='Jumlah',
        titlefont_size=16,
        tickfont_size=14))
barplot.show()
```
