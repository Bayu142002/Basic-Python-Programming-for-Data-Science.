#Nomor 1

```python
import numpy as np
array = []
i = int(input("Elemen array: "))
for x in range(i):
  data = float(input("{} = ".format(x+1)))
  array.append(data)
  if data > 0 :
    print('{} Saldo deposit'.format(data))
  elif data == 0:
    print('{} Saldo zero'.format(data))
  else:
    print('{} Saldo withdraw'.format(data))
print("\nRata-rata saldo = {}".format(sum(array)/i))
```
