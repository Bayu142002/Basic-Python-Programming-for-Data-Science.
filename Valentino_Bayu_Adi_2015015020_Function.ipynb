# 1. Mencari Luas Lingkaran

```python
def luas_lingkaran (pi,r):
  L_ling = pi * (r^2)
  return L_ling
```

```python
luas_lingkaran(3.14,21)
```

# 2. Mencari Luas Segitiga

```python
def luas_segitiga (alas,tinggi):
  L_seg = 1/2*(alas*tinggi)
  return L_seg
```

```python
luas_segitiga(12,8)
```

```python
def penyelesaian(nomor, x, y):
  print("x : sebagai pi dan alas")
  print("y : sebagai jari-jari dan tinggi\n")
  if nomor == 1:
    print("Luas Lingkaran")
    hasil = x * (y^2)
  elif nomor == 2:
    print("Luas Segitiga")
    hasil = 1/2*(x*y)
  else:
    print("mencari nilai max")
    x = int(input())
    y = int(input())
    z = int(input())
    hasil = print("nilai max = ",max(x,y,z))
  return hasil
  
```
